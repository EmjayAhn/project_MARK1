```python
response = TextResponse(req.url, body=req.text, encoding='utf-8')
```

```python
trademarks = response.xpath('//*[@id="container"]/ul')
trademarks
```

```python
items_row = trademarks[0].xpath('./li')
items_row[0].xpath('./a/span/img/@src').extract()[0]
```
